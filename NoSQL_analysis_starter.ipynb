### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene' : 20}

# Use count_documents to display the number of documents in the result
hygiene_count = establishments.count_documents(query)

# Turning hygiene_results into list and finding all documents that match based on the criteria above in "query"   
hygiene_results = list(establishments.find(query)) 

# Printing an f string to show the number of documents returned based on the query above
print(f'Number of documents with a Hygiene score of 20 in the "uk_food" database: {hygiene_count}')

# Display the first document in the results using pprint ==> [:1] grabs the first document due to "lond_rating_results" being converted into a list above
for doc in hygiene_results[:1]:
    pprint(doc)
```

```python
# Convert the result to a Pandas DataFrame ==> using "pd.json_normalize" instead of "pd.DataFrame" to split up nested dictionaries inside of the overall document into individual columns
hygiene_df = pd.json_normalize(hygiene_results)

# Display the number of rows in the DataFrame
print(f'Number of rows in the hygiene_df : {len(hygiene_df)}')

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

# Retrieving establishments that have "London" in the name for "LocalAuthorityName" and a "RatingValue" of greather than or equal to 4 ==> '$regex' finds all "LocalAuthorityName" that have "London" in the name
query = {'LocalAuthorityName': {'$regex': 'London'},
            'RatingValue' : {'$gte' : 4}}

# Use count_documents to display the number of documents in the result
lond_rating_counts = establishments.count_documents(query)


# Turning lond_rating_results into list and finding all documents that match based on the criteria above in "query"
lond_rating_results = list(establishments.find(query))

# Printing an f string to show the number of documents returned based on the query above
print(f'Number of documents in London with a RatingValbe greater than or equal to 4 in the "uk_food" database: {lond_rating_counts}')

# Display the first document in the results using pprint ==> [:1] grabs the first document due to "lond_rating_results" being converted into a list above
for results in lond_rating_results[:1]:
    pprint(results)
```

```python
# Convert the result to a Pandas DataFrame ==> using "pd.json_normalize" instead of "pd.DataFrame" to split up nested dictionaries inside of the overall document into individual columns
london_df = pd.json_normalize(lond_rating_results)

# Display the number of rows in the DataFrame
print(f'Number of rows in the london_df : {len(london_df)}')

# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Defining variables for calculations in "query"
degree_search = 0.01
latitude = 51.49014200      # ==> Lat of "Penang Flavours"
longitude = 0.08384000      # ==> Long of "Penang Flavours"


# Used to retrieve the establishments that are within 0.01 degree on either side of the latitude and longitude AND establishments that have a "RatingValue" equal to 5
query = {"geocode.latitude": {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
         "geocode.longitude": {'$gte': longitude - degree_search, '$lte': longitude + degree_search},
         "RatingValue": 5}

# Sorting by the Hygiene score in ascending order ==> "1"
sort = [('scores.Hygiene', 1)]

# Limiting the results to 5
limit = 5

# Finding the documents/"establishments" that fit the requirments from "query" ==> Then sorting by hygiene score ==> Then limiting the results to only the top 5 establishments
top_five_results = list(establishments.find(query).sort(sort).limit(limit))

# Print the results
for result in top_five_results:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame ==> using "pd.json_normalize" instead of "pd.DataFrame" to split up nested dictionaries inside of the overall document into individual columns
top_five_df = pd.json_normalize(top_five_results)
top_five_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# MATCH STAGE for establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 0}}

# GROUP STAGE to group by "LocalAuthorityName"
group_query = {'$group': {'_id': "$LocalAuthorityName", 'count': {'$sum': 1}}}

# SORT STAGE to sort matches from highest to lowest (-1 => Descending order => AKA: Highest to Lowest)
sort_query = {'$sort': {'count': -1}}

# Creating the pipeline of queries/"stages"
pipeline = [match_query, group_query, sort_query]

# Using the ".aggregate" function on the pipeline to query and calculate the results
pipeline_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result "pipeline_results"
print(f'Number of documents in the pipeline_results: {len(pipeline_results)}')

# Print the first 10 results
for result in pipeline_results[:10]:
    pprint(result)
```

```python
# Convert the result to a Pandas DataFrame  => ('pd.DataFrame' is sufficent for these results because there is no need to split up nested dictionaries based on the output that is given from the pipeline query)
pipeline_df = pd.DataFrame(pipeline_results)

# Display the number of rows in the DataFrame
print(f'Number of rows in the london_df : {len(pipeline_df)}')

# Display the first 10 rows of the DataFrame
pipeline_df.head(10)
```
