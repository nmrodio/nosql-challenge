### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene' :{'$eq' : 20}}

# Use count_documents to display the number of documents in the result
hygiene_count = establishments.count_documents(query)

# Display the first document in the results using pprint   
hygiene_results = list(establishments.find(query))                #==========>   ### I THINK LIST SHOULD GO HERE TO MAKE THE DATAFRAME !!!!!!

# !!!!!!!!!!!!!!!!!!!!!!!!!!!! DO THEY WANT THIS LITERALLY PRINTED OUT !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
print(f'Number of documents with a Hygiene score of 20 in the "uk_food" database: {hygiene_count}')

for doc in hygiene_results[:1]:
    pprint(doc)
```

```python
# Convert the result to a Pandas DataFrame
#hygiene_df = pd.DataFrame(hygiene_results)
hygiene_df = pd.json_normalize(hygiene_results)
# Display the number of rows in the DataFrame
print(f'Number of rows in the hygiene_df : {len(hygiene_df)}')

# Display the first 10 rows of the DataFrame
hygiene_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Match Stage for establishments with a hygiene score of 0
match_stage = {'$match': {'scores.Hygiene': {'$eq': 0}}}

# GROUP STAGE to group by "LocalAuthorityName"
group_stage = {'$group': {'_id': "$LocalAuthorityName", 'count': {'$sum': 1}}}

# SORT STAGE to sort matches from highest to lowest (-1 => Descending order => AKA: Highest to Lowest)
sort_stage = {'$sort': {'count': -1}}

# Creating the pipeline of queries
pipeline = [match_stage, group_stage, sort_stage]

# Using the "".aggregate" function on the pipeline to query and calculate the results
pipeline_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f'Number of documents in the pipeline_results: {len(pipeline_results)}')

# Print the first 10 results
for result in pipeline_results[0:10]:
    pprint(results)
```

```python
# Convert the result to a Pandas DataFrame
pipeline_df = pd.json_normalize(pipeline_results)
# Display the number of rows in the DataFrame
print(f'Number of rows in the london_df : {len(london_df)}')
# Display the first 10 rows of the DataFrame
pipeline_df.head(10)
```
